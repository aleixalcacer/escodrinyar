```python
import escodrinyar as sc
import seaborn.objects as so
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
points = (
    sc.Plot(
        data=penguins,
        x="flipper_length_mm",
        y="bill_length_mm",
        color="species"
    )
    .add(so.Dot())
)

chull = (
    sc.Plot(
        data=penguins,
        x="flipper_length_mm",
        y="bill_length_mm",
        color="species"
    )
    .add(sc.ConvexHull(edgewidth=1))
)
```

```python
points + chull
```

```python
(points + chull).opts(
    figsize=(10, 5)
)
```
